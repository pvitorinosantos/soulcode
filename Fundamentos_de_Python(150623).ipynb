### Estrutura de Condição - if, else, elif

Instrução IF:


*   A instrução IF é usada para executar um bloco de códigos se uma determinada condição for verdadeira. A sintaxe geral é a seguinte:



```python
'''
if condição:
  # bloco de código a ser executado se a condição for verdadeira
  '''
```

```python
idade = 19
if idade >=18:
  print('Você é maior de idade')
```

```python
idade = 17
if idade >=18:
  print('Você é maior de idade')
```

Instrução ELSE:


1.   A instrução ELSE é usada em conjunto com a instrução IF e especifica um bloco de código a ser executado se a condição do IF não for verdadeira. A sintaxe geral é a seguinte:



```python
'''
if condição:
  # bloco de código a ser executado se a condição for verdadeira
else:
  # bloco de código a ser executado se a condição não for verdadeira
  '''
```

```python
idade = int(input('Digite sua idade: '))

if idade >= 18:
  print('Você é maior de idade')
else:
  print('Você é menor de idade')
```

Instrução ELIF


*   A instrução ELIF é uma abreviação de "ELSE IF" e permite testar múltiplas condições em sequência. Ela é usada quando há mais de duas possibilidades a serem consideradas. A sintaxe geral é a seguinte:



```python
'''
if condição1:
  # bloco de código a ser executado se a condição 1 for verdadeira
elif condição2:
  # bloco de código a ser executado se a condição 2 for verdadeira
else:
  # bloco de código a ser executado se nenhuma das condições anteriores for verdadeira
  '''
```

```python
nota = float(input('Digite sua nota: '))

if nota >=90:
  print('Sua nota é A')
elif nota >=80:
  print('Sua nota é B')
elif nota >=70:
  print('Sua nota é C')
else:
  print('Sua nota é D')
```

```python
# Lista é uma estrutura que pode armazenar diversos valores sobre um determinado atributo
lista_vazia =  []
noivo = ['Zé', 'Mané', 'Ana', 'Franciane', 'João', 'Franz']
print(f'Convidados do noivo - {noivo}')
idades = [41, 47, 46, 28, 17, 50]
print(f'Idade dos convidados do noivo = {idades}')
massas = [86, 74, 52, 47, 90, 110]
alturas = [1.68, 1.74, 1.65, 1.65, 1.80, 1.97]
print(f'Massas dos convidados do noivo = {massas}')
print(f'Alturas dos convidados do noivo = {alturas}')
```

```python
# Acessando um elemento específico da lista a partir do seu índice
print(f'A convidade de honra é {noivo[3]}')
# Criar uma sublista com apenas os três primeiros elementos da lista noivo
sublista = noivo[:3]
print(sublista)
print(f'O primeiro convidado é {noivo[0]} e o último convidado é {noivo[-1]}')
print(f'{noivo}')
```

```python
# Método index (valor procurado, índice de início da procura e índice do fim da procura) por definição se não for oferecido o índice
id=noivo.index('Franciane')
```

```python
# Qual é o IMC da convidada Fran?
print(f'O IMC da Fran é {(massas[id]/alturas[id]**2):.2f}')
```

Desafio 1: Crir um programa que receba do usuário um nome e caso esse nome esteja na lista, guarde seu índice em uma variável e a utilize buscar massa e altura correspondente, bem como calcular seu IMC e após tudo isso informar o usuário. Caso o nome não esteja na lista, retorne uma mensagem informando o usuário,

```python
nome = str(input('Digite o nome que deseja consultar: ')) # Recebendo o nome do usuário
noivo = ['Zé', 'Mané', 'Ana', 'Fran', 'João', 'Franz'] # Cria o banco de dados com os nomes dos convidados do noivo
nome = nome.capitalize() # Deixo o nome que o usuário digitou com a primeira letra em caixa alta e as demais em caixa baixa
if nome in noivo: #se o nome que o usuário digitou estiver no banco de dados, faça algo
  id2 = noivo.index(nome) # Recebendo o argumento nome e procurando o seu índice, já sabendo que ele está na lista exatamente como digitado
  imc = massas[id2]/alturas[id2]**2 # Calculando o IMC já sabendo a posição que deve buscar em cada lista (massas e alturas)
  print(f'{nome} tem {massas[id2]} kg, {alturas[id2]} metro(s) e seu IMC é: {imc:.1f}')
else:
  print(f'Desculpe, {nome} não está na lista. Por favor, reinicie o programa para consultar outro nome')
```

```python
#  Adicionando elementos em uma lista qualquer com o método append (). Por padrão, este método insere o valor desejado em uma lista existente na última posição
lista_vazia = [] # Cria uma lista vazia com o nome lista vazia e apaga todos os elementos de qualquer objeto com o mesmo nome
print(lista_vazia) # Imprime uma lista vazia que já existe
lista_vazia.append('Joana') # Adiciona Joana na última posição de uma lista que já exista
lista_vazia.append('Douglas')
lista_vazia.append('Eduardo')
lista_vazia.append('Ester')
lista_vazia.append('Daniela')
lista_vazia.append('Pedro')
lista_vazia.append('Maria')
print(lista_vazia) # Imprime o status atual da lista vazia
fam = lista_vazia
print(lista_vazia)
print(fam)
```
